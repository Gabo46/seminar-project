```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot")

factor = [0, 2, 4, 8, 16]
comp = [0, 30, 60, 75, 85, 95]
```

```python
def plot(dictc, xval, n, title):
    for key in dictc.keys():
        y = np.array(dictc[key][n])
        y /= y[-1]
        plt.scatter(xval, y)
        name = key.capitalize()
        plt.plot(xval, y, label=name)
    plt.title(title)
    plt.xlabel("Komprimierungsfaktor" if n == 1 else "Skalierungsfaktor")
    plt.ylabel("Skala")
    plt.legend()
```

```python
fig = plt.figure(figsize=(18,3.6))
plt.subplot(131)
plot(cartoon, comp, 1, title="Cartoon")
plt.subplot(132)
plot(face, comp, 1, title="Gesicht")
plt.subplot(133)
plot(landscape, comp, 1, title="Landschaft")
```

```python
fig = plt.figure(figsize=(18,3.6))
plt.subplot(131)
plot(cartoon, factor, 0, title="Cartoon")
plt.subplot(132)
plot(face, factor, 0, title="Gesicht")
plt.subplot(133)
plot(landscape, factor, 0, title="Landschaft")
```

```python
def plot2(dicts, names, key, xvals, n):
    plt.figure(figsize=(5,3))
    for i, dic in enumerate(dicts):
        y = np.array(dic[key][n]) 
        y /= y[-1]
        plt.scatter(xvals, y)
        name = names[i]
        plt.plot(xvals, y, label=name)
    plt.title(key.capitalize())
    plt.xlabel("Komprimierungsfaktor" if n == 1 else "Skalierungsfaktor")
    plt.ylabel("Skala")
    plt.legend()
```

```python
plot2([face, landscape, cartoon], ["Gesicht", "Landschaft", "Cartoon"], "gabriel", factor, 0)
```
