# Horizontal Cuts for the images

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec
from PIL import Image
plt.style.use("ggplot")
```

```python
def show_horizontal_cut(im):
    """ Shows an horizontal luminance profiles across the middle of 
    one 8-bit grayscale image
    """
    midrow = int(im.shape[0]/2)
    plt.figure()
    plt.plot(im[midrow,:, 0], linewidth=1, color="red")
    plt.plot(im[midrow,:, 1], linewidth=1, color="green")
    plt.plot(im[midrow,:, 2], linewidth=1, color="blue")
    plt.show()
```

```python
def add_horizontal_cut(im, ax):
    """ Shows an horizontal luminance profiles across the middle of 
    one 8-bit grayscale image
    """
    midrow = int(im.shape[0]/2)
    ax.plot(im[midrow,:, 0], linewidth=1, color="red")
    ax.plot(im[midrow,:, 1], linewidth=1, color="green")
    ax.plot(im[midrow,:, 2], linewidth=1, color="blue")
```

## Landscape

```python
landscape_95 = np.array(Image.open("images/landscape/95_scaled.jpg"))
```

```python
fontdict = {'fontsize': 10}
```

```python
fig2 = plt.figure(figsize=(20,20))
spec2 = GridSpec(ncols=4, nrows=3, figure=fig2)

f2_ax1 = fig2.add_subplot(spec2[0, :2])
f2_ax1.set_title("Originalbild", fontsize=20)
f2_ax1.set_xlabel("Pixel")
f2_ax1.set_ylabel("Intensitt")

add_horizontal_cut(landscape_original, f2_ax1)
f2_ax2 = fig2.add_subplot(spec2[0, 2:])
f2_ax2.set_title("Skalierungsstufe 2", fontsize=20)
f2_ax2.set_xlabel("Pixel")
f2_ax2.set_ylabel("Intensitt")

add_horizontal_cut(landscape_x2, f2_ax2)
f2_ax3 = fig2.add_subplot(spec2[1, :2])
add_horizontal_cut(landscape_x4, f2_ax3)
f2_ax3.set_title("Skalierungsstufe 4", fontsize=20)
f2_ax3.set_xlabel("Pixel")
f2_ax3.set_ylabel("Intensitt")

f2_ax4 = fig2.add_subplot(spec2[1, 2:])
add_horizontal_cut(landscape_x8, f2_ax4)
f2_ax4.set_title("Skalierungsstufe 8", fontsize=20)
f2_ax4.set_xlabel("Pixel")
f2_ax4.set_ylabel("Intensitt")

f2_ax5 = fig2.add_subplot(spec2[2, 1:3])
add_horizontal_cut(landscape_x16, f2_ax5)
f2_ax5.set_title("Skalierungsstufe 16", fontsize=20)
f2_ax5.set_xlabel("Pixel")
f2_ax5.set_ylabel("Intensitt")

fig2.subplots_adjust(wspace=.4)
fig2.subplots_adjust(hspace=.2)

plt.show()
```
